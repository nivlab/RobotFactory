```python
def get_data(results_name):
    with gzip.open('../stan_results/{}.tsv.gz'.format(results_name), 'rb') as f_in:
        with open('../stan_results/{}.tsv'.format(results_name), 'wb') as f_out:
            shutil.copyfileobj(f_i
                               n, f_out)

    fit_summ = read_csv('../stan_results/{}_summary.tsv'.format(results_name), 
             sep='\t')
    samples=read_csv('../stan_results/{}.tsv'.format(results_name), 
             sep='\t')
    
    return fit_summ, samples

def plot_eta_beta(samples, n_params=2, c=10):
    names = ['beta', 'eta', 'eta']
    for param in range(n_params):
        plt.subplot(1, n_params, param+1)
        plt.title(names[param])
        if names[param]=='beta':
            plt.hist(norm.cdf(samples['theta_mu[{}]'.format(param+1)])*c)
        else:
            plt.hist(norm.cdf(samples['theta_mu[{}]'.format(param+1)]))
    plt.tight_layout()
    plt.show()
```

```python
# [['theta_mu[1]','theta_mu[2]',
#                 'theta_mu[3]','theta_mu[4]',
#                 'theta_mu[5]','theta_mu[6]']]
samples_sub_6e_theta_melted = samples_sub_6e_theta.melt("beta",
                                                        var_name='all_eta',
                                                        value_name="val")
samples_sub_6e_theta_melted['all_eta'] = samples_sub_6e_theta_melted['all_eta'].astype(int)
sns.scatterplot(y='beta', x="val",
                palette="ch:r=-.2,d=.3_r",
                hue='all_eta',
               # sizes=(1, 8), linewidth=0,
                data=samples_sub_6e_theta_melted)
```

```python
sns.boxplot( x="all_eta",y='val',
                palette="ch:r=-.2,d=.3_r",
                #sizes=(1, 8), linewidth=0,
                data=samples_sub_6e_theta_melted)

sns.regplot(x="all_eta", y='val',
            data=samples_sub_6e_theta_melted,
            ci=68,
            scatter=False)
```
